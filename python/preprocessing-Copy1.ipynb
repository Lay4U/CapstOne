```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('dataset_origin.csv')
del df['Unnamed: 0']
df
```

```python
df.name = df.name.str.replace(r'\(\w*\)', '')
df
```

```python
df.name = df.name.str.replace('동대문운동장', '동대문역사문화공원')
df.name = df.name.str.replace('동대문역사문화공원5', '동대문역사문화공원')
df.name = df.name.str.replace('성내', '잠실나루')
df.name = df.name.str.replace('신천', '잠실새내')
df.name = df.name.str.replace('미아삼거리', '미아사거리')
df.name = df.name.str.replace('이수', '총신대입구')
df
```

```python
s = df['name']
s = list(sorted(set(s)))
s
```

```python
df2 = df.copy()
for time in ['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t']:
    df2[time] = df2[time] - df2[time].shift(-1)
df2
```

```python
df2 = df2[::2]
df2
```

```python
df3 = df2.melt(id_vars=df.columns.tolist()[0:4], value_name='a').drop('variable', 1)

df3 = df3.sort_values(['date','number', 'name'], ascending=[True, True, False])
df3
```
