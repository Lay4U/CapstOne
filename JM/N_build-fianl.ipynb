```python
#데이터 역명 불러오기
s=pd.read_csv('outputFinal.csv',encoding='cp949')
s
```

```python
s=s['name']
```

```python
type(s)
s=list(s)
for i in range(len(s)):
    s[i]=s[i].strip()
s
```

```python
list1=list()
for i in range(len(s)):
    url="http://swopenapi.seoul.go.kr/api/subway/504f484c4673616c39304369624854/xml/gateInfo/0/30/"+s[i]
    result=requests.get(url)
    bs_obj=BeautifulSoup(result.content,"html.parser")
    bd=bs_obj.find_all("cfrbuild")
    bd=str(bd)
    num=len(bd.split(","))
    list1.append(num)
```

```python
data={"name":s,"building":list1}
BN=pd.DataFrame(data)
BN
```
