```python
#데이터 역명 불러오기
s=pd.read_csv('StationName.csv')
del s['Unnamed: 1']
s=s.dropna()
```

```python
s=s['name']
```

```python
s=list(s)
s
```

```python
#한글만 추출하여 역명 전처리

for i in range(len(s)):
    t = re.sub('{|\'|\'','', s[i])
    t = t.split("(")
    s[i]= t[0]
    s[i]=s[i].strip()
```

```python
s=list(set(s))
s
```
